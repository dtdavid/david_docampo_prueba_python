3. Agrega las siguientes columnas, considerando su nombre y la fórmula asociada
- **venta**: quantityOrdered * priceEach
- **costo**: quantityOrdered * buyPrice
- **ganancia**: considerando las columnas anteriores

```python
# comprobamos el tipo de dato que tiene cada columna
print(df_merged['quantityOrdered'].dtypes) 
print(df_merged['priceEach'].dtypes)
print(df_merged['buyPrice'].dtypes)
#corregimos cualquier dato nulo y lo pasamos a 0 numérico
df_merged['quantityOrdered'] = df_merged['quantityOrdered'].fillna(0)
df_merged['priceEach'] = df_merged['priceEach'].fillna(0)
df_merged['buyPrice'] = df_merged['buyPrice'].fillna(0)
#creamos las nuevas columnas basándonos en las fórmulas que nos facilitan
df_merged['venta'] = df_merged['quantityOrdered'] * df_merged['priceEach']
df_merged['costo'] = df_merged['quantityOrdered'] * df_merged['buyPrice']
df_merged['ganancia'] = df_merged['venta'] - df_merged['costo']
print(df_merged[['venta','costo' ,'ganancia']])
display((df_merged).head())
```

4. ¿Cuál fue el total de ventas por línea de productos? Incluye una fila de totales.

```python
# tendremos en cuenta la columna df_merged['productLine'] y df_merged['venta']
print(df_merged[['productLine','venta']].head())
# nos pide el total de ventas 'sum()' por 'productLine'
# e incluir una fila de totales 'All'
ventas_por_linea = df_merged.groupby('productLine')['venta'].sum().reset_index()
# en pandas primero agrupamos .groupby('productLine')
# después aplicamos .sum() a la columna ['venta']
# .reset_index() esto es para impedir que 'productLine' sea el índice
print(f"\n{ventas_por_linea}\n")
# falta añadir una columna de totales 'All'
# guardamos en una variable el total de ventas ['ventas'].sum()
# partiendo de la fórmula pd.DataFrame({'columna1': [valor1], 'columna2': [valor2]})
# y después concatenamos con ventas_por_linea
totales = pd.DataFrame({'productLine': ['All'], 'venta': [ventas_por_linea['venta'].sum()]})
ventas_con_totales = pd.concat([ventas_por_linea, totales], ignore_index=True) # reseteamos los índices
print(f"Total de ventas por línea de productos:\n{ventas_con_totales}")
```

5. ¿Cuántos clientes distintos hicieron compras?

```python
# qué clientes hicieron compras. Usamos la columna 'status'
# sacar de clientes que hicieron compras, una lista sin duplicados, con método .unique()
# después usar len() para que nos dé el número total

clientes_con_compras = df_merged[df_merged['status'] == 'Shipped']
clientes_unicos = clientes_con_compras['customerNumber'].unique()
cantidad_clientes_unicos = len(clientes_unicos)
print(f"Clientes distintos que compraron: {cantidad_clientes_unicos}")
#(df_merged['status']).unique()
```

6. ¿Existen clientes que aún no han hecho ninguna compra? ¿Cuántos son?

```python
# qué clientes NO hicieron compras. Usamos la columna 'status'
# sacar de clientes que NO hicieron compras, una lista sin duplicados, con método .unique() 
# después usar len() para que nos dé el número total
# finalmente podemos sacar el listado completo de clientes, para comprobar si la suma de clientes_con_compras + clientes_sin_compras = total_clientes

# Clientes con compras
clientes_con_compras = df_merged[df_merged['status'] == 'Shipped']['customerNumber'].unique()

# Clientes sin compras
clientes_sin_compras = df_merged[~df_merged['customerNumber'].isin(clientes_con_compras)]['customerNumber'].unique() # aquí buscamos la negación de clientes_con_compras anteponiendo ~ 

# Total de clientes únicos
total_clientes = df_merged['customerNumber'].nunique()

# Contar clientes
total_con_compras = len(clientes_con_compras)
total_sin_compras = len(clientes_sin_compras)

# Mostrar resultados
print(f"Clientes con compras: {total_con_compras}")
print(f"Clientes sin compras: {total_sin_compras}")
print(f"Total clientes únicos: {total_clientes}")

print('¿Existen clientes que aún no han hecho ninguna compra? ¿Cuántos son?  ')
if total_sin_compras == 0:
    print(f'No. Los clientes sin compras son {total_sin_compras}')
else:
    print(f"SI. En total hay: {total_sin_compras}")

# por los resultados, es posible que todos los clientes hayan comprado y que 
# también estén en otro 'status' 
status = df_merged['status'].unique()
print(f"\nDiferentes tipos de estado: ")
for estado in status:
    print(estado)
# 'Shipped', 'Resolved', 'Cancelled', 'On Hold', 'Disputed', 'In Process'
       
```
